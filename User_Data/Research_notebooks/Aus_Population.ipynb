```python
# Seasonal Decomposition and Deseasonalization Functions
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings

def decompose_and_deseasonalize(data, method='additive', model_type='multiplicative', 
                               period=None, extrapolate_trend=0, return_components=False):
    """
    Decompose time series and remove seasonality using statsmodels.
    
    Parameters:
    -----------
    data : pandas.Series
        Time series data with datetime index
    method : str, default 'additive'
        Type of seasonal decomposition ('additive' or 'multiplicative')
    model_type : str, default 'multiplicative' 
        How to remove seasonality ('additive' or 'multiplicative')
    period : int, optional
        Seasonal period. If None, will attempt to infer from data frequency
    extrapolate_trend : int, default 0
        Number of periods to extrapolate trend at beginning and end
    return_components : bool, default False
        If True, returns decomposition components along with deseasonalized data
        
    Returns:
    --------
    pandas.Series or tuple
        Deseasonalized time series, optionally with decomposition components
    """
    
    # Ensure data has datetime index
    if not isinstance(data.index, pd.DatetimeIndex):
        raise ValueError("Data must have a datetime index")
    
    # Remove any NaN values
    clean_data = data.dropna()
    
    # Ensure we have enough data points for decomposition
    if len(clean_data) < 2:
        warnings.warn("Not enough data points for decomposition. Returning original data.")
        if return_components:
            return data, None
        return data
    
    # Infer period if not provided
    if period is None:
        freq = pd.infer_freq(clean_data.index)
        if freq:
            if 'M' in freq:
                period = 12  # Monthly data
            elif 'Q' in freq:
                period = 4   # Quarterly data
            elif 'D' in freq:
                period = 365 # Daily data (approximate)
            else:
                period = 12  # Default fallback
        else:
            period = 12  # Default fallback
    
    # Ensure we have enough periods for decomposition
    if len(clean_data) < 2 * period:
        warnings.warn(f"Not enough data points ({len(clean_data)}) for period {period}. Using available data length.")
        period = max(2, len(clean_data) // 2)
    
    # Perform seasonal decomposition
    try:
        decomposition = seasonal_decompose(clean_data, 
                                         model=method, 
                                         period=period,
                                         extrapolate_trend=extrapolate_trend)
        
        # Remove seasonality
        if model_type == 'additive':
            deseasonalized = clean_data - decomposition.seasonal
        else:  # multiplicative
            deseasonalized = clean_data / decomposition.seasonal
        
        # Reindex to match original data index (including NaN positions)
        deseasonalized = deseasonalized.reindex(data.index)
        
        if return_components:
            return deseasonalized, decomposition
        else:
            return deseasonalized
            
    except Exception as e:
        warnings.warn(f"Seasonal decomposition failed: {str(e)}. Returning original data.")
        if return_components:
            return data, None
        return data

def plot_decomposition(data, method='additive', period=None, title="Time Series Decomposition"):
    """
    Plot seasonal decomposition components including the deseasonalized series.
    
    Parameters:
    -----------
    data : pandas.Series
        Time series data with datetime index
    method : str, default 'additive'
        Type of seasonal decomposition ('additive' or 'multiplicative')
    period : int, optional
        Seasonal period
    title : str
        Plot title
    """
    
    deseasonalized, decomp = decompose_and_deseasonalize(data, method=method, 
                                                        period=period, 
                                                        return_components=True)
    
    if decomp is None:
        print("Decomposition failed. Plotting original data only.")
        fig, ax = plt.subplots(1, 1, figsize=(12, 4))
        ax.plot(data.index, data.values)
        ax.set_title(f'{title} - Original Data Only')
        ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        return fig, ax
    
    fig, axes = plt.subplots(5, 1, figsize=(12, 12))
    fig.suptitle(title, fontsize=16)
    
    # Original
    axes[0].plot(data.index, data.values, color='blue')
    axes[0].set_title('Original')
    axes[0].grid(True, alpha=0.3)
    
    # Trend
    axes[1].plot(decomp.trend.index, decomp.trend.values, color='green')
    axes[1].set_title('Trend')
    axes[1].grid(True, alpha=0.3)
    
    # Seasonal
    axes[2].plot(decomp.seasonal.index, decomp.seasonal.values, color='orange')
    axes[2].set_title('Seasonal Component')
    axes[2].grid(True, alpha=0.3)
    
    # Deseasonalized (this was missing!)
    axes[3].plot(deseasonalized.index, deseasonalized.values, color='red')
    axes[3].set_title('Deseasonalized')
    axes[3].grid(True, alpha=0.3)
    
    # Residual
    axes[4].plot(decomp.resid.index, decomp.resid.values, color='purple')
    axes[4].set_title('Residual')
    axes[4].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return fig, axes
```
